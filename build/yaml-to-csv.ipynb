```python
import csv
import yaml
import pandas as pd
```

```python
def yaml_to_df(filename):
    f = open('../master/{filename}.yaml'.format(filename = filename), newline='')
    data = yaml.load(f, Loader=yaml.FullLoader)
    f.close()
    return pd.DataFrame.from_dict(data)

def yaml_to_csv(filename):
    df = yaml_to_df(filename)
    df.to_csv('{filename}.csv'.format(filename=filename), index=False)
```

## Terms defined in TCS

```python
filename = 'tcs'

df = yaml_to_df(filename)
yaml_to_csv(filename)
df
```
