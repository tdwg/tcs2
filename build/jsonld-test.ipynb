```python
from pyld import jsonld
import json

expanded = jsonld.expand('https://raw.githubusercontent.com/tdwg/tcs2/examples/docs/examples/example1/example1.jsonld')
print(json.dumps(expanded, indent=2))
```

```python
compacted = jsonld.compact(expanded, 'https://raw.githubusercontent.com/tdwg/tcs2/examples/docs/examples/context.jsonld')
print(json.dumps(compacted, indent=2))
```
