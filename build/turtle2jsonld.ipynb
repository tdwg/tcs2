```python
from rdflib import Graph
from pyld import jsonld

context = "https://raw.githubusercontent.com/tdwg/tcs2/master/examples/context.jsonld"

with open('../recipes/giraffes.ttl', 'r') as file:
    turtle = file.read()

    g = Graph().parse(data=turtle, format='turtle')
    with open('../recipes/giraffes.jsonld', 'w') as f:
        expanded = g.serialize(format='json-ld')
        compacted = jsonld.compact(expanded, context)
        f.write(compacted)
```
